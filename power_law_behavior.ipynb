```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def power_law_function(x, a, k):
    """
        A best-fit model that follows the power law distribution: y = a * x^(-k),
        where a and k are fitting parameters.

        Args:
            x (np.array): Array of x values
            a (float): intercept or scale factor/constant (shifts curve up/down). It does not affect the shape/decay rate of the curve (only how high the curve starts)
            k (float): exponent of the power law distribution. It defines how fast y decays as x increases.
    """
    return a * x ** (-k)

```

```python
# we define x 100 values from a range of 0-1000
x_vals = np.linspace(0, 1000, 100)
```

```python
# a behaviors
a = 5
```

```python
# k < -1
"""
y increases exponentially as x increases:	Not a power law decay — reverse power law
"""

# k < 0
k1 = -3
k2 = -2.4
k3 = -2

y1 = power_law_function(x_vals, a, k1)
y2 = power_law_function(x_vals, a, k2)
y3 = power_law_function(x_vals, a, k3)

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1, 1, 1)
#ax.plot(x_vals, y1, label=f"k1 = {k1}")
ax.plot(x_vals, y2, label=f"k2 = {k2}")
ax.plot(x_vals, y3, label=f"k3 = {k3}")
ax.legend()
```

```python
# k < 0
"""
y increases as x increases:	Not a power law decay — reverse power law
"""

k1 = -1
k2 = -0.6
k3 = -0.24
k4 = -0.02

y1 = power_law_function(x_vals, a, k1)
y2 = power_law_function(x_vals, a, k2)
y3 = power_law_function(x_vals, a, k3)
y4 = power_law_function(x_vals, a, k4)

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1, 1, 1)
#ax.plot(x_vals, y1, label=f"k1 = {k1}")
ax.plot(x_vals, y2, label=f"k2 = {k2}")
ax.plot(x_vals, y3, label=f"k3 = {k3}")
ax.plot(x_vals, y4, label=f"k4 = {k4}")
ax.legend()
```

```python
# k == 0
"""
y=a: a constant	Flat horizontal line
"""
k = 0
y = power_law_function(x_vals, a, k)

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1, 1, 1)
ax.plot(x_vals, y, label=f"k = {k}")
ax.legend()
```

```python
# 0 < k < 1
"""
Very slow decay	Long "tail": almost flat; heavy tail
"""
k1 = 0.02
k2 = 0.4
k3 = 0.8
k4 = 0.99

y1 = power_law_function(x_vals, a, k1)
y2 = power_law_function(x_vals, a, k2)
y3 = power_law_function(x_vals, a, k3)
y4 = power_law_function(x_vals, a, k4)

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1, 1, 1)
ax.plot(x_vals, y1, label=f"k1 = {k1}")
ax.plot(x_vals, y2, label=f"k2 = {k2}")
ax.plot(x_vals, y3, label=f"k3 = {k3}")
ax.plot(x_vals, y4, label=f"k4 = {k4}")
ax.legend()
```

```python
# k == 1
"""
y∼1/x:	Zipf's law, inverse law
"""
k = 1
y = power_law_function(x_vals, a, k)

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1, 1, 1)
ax.plot(x_vals, y, label=f"k = {k}")
ax.legend()
```

```python
# 1 < k < 2
"""
Faster decay, infinite variance but finite mean:	Found in wealth, city sizes, web traffic
"""
k1 = 1.2
k2 = 1.65
k3 = 1.99

y1 = power_law_function(x_vals, a, k1)
y2 = power_law_function(x_vals, a, k2)
y3 = power_law_function(x_vals, a, k3)

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1, 1, 1)
ax.plot(x_vals, y1, label=f"k1 = {k1}")
ax.plot(x_vals, y2, label=f"k2 = {k2}")
ax.plot(x_vals, y3, label=f"k3 = {k3}")
ax.legend()
```

```python
# k == 2
"""
y∼1/sqr(x) Borderline: mean and variance can diverge
"""
k = 2
y = power_law_function(x_vals, a, k)

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1, 1, 1)
ax.plot(x_vals, y, label=f"k = {k}")
ax.legend()
```

```python
# k > 2
"""
Rapid decay, lighter tail:	Data becomes less "fat-tailed"
"""
k1 = 2.21
k2 = 2.46
k3 = 2.5
k4 = 2.78
k5 = 2.9

y1 = power_law_function(x_vals, a, k1)
y2 = power_law_function(x_vals, a, k2)
y3 = power_law_function(x_vals, a, k3)
y4 = power_law_function(x_vals, a, k4)
y5 = power_law_function(x_vals, a, k5)

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1, 1, 1)
ax.plot(x_vals, y1, label=f"k1 = {k1}")
ax.plot(x_vals, y2, label=f"k2 = {k2}")
ax.plot(x_vals, y3, label=f"k3 = {k3}")
ax.plot(x_vals, y4, label=f"k4 = {k4}")
ax.plot(x_vals, y4, label=f"k4 = {k5}")
ax.legend()
```

```python
# k > 3
"""
Decay resembles exponential:	Tail becomes too light for typical power law
"""
k1 = 3
k2 = 3.6
k3 = 4

y1 = power_law_function(x_vals, a, k1)
y2 = power_law_function(x_vals, a, k2)
y3 = power_law_function(x_vals, a, k3)

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1, 1, 1)
ax.plot(x_vals, y1, label=f"k1 = {k1}")
ax.plot(x_vals, y2, label=f"k2 = {k2}")
ax.plot(x_vals, y3, label=f"k3 = {k3}")
ax.legend()
```
